```python
import pandas as pd

uri = "https://gist.githubusercontent.com/guilhermesilveira/b9dd8e4b62b9e22ebcb9c8e89c271de4/raw/c69ec4b708fba03c445397b6a361db4345c83d7a/tracking.csv"
dados = pd.read_csv(uri)
dados.head()
```

```python
y = dados["comprou"]
y.head()
```

```python
x = dados[["inicial",	"palestras",	"contato",	"patrocinio"]]
x.head()
```

Quantas linhas no meu arquivo?

```python
dados.shape
```

Treinando o modelo

```python
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score

treino_x = x[:75]
treino_y = y[:75]

teste_x = x[75:]
teste_y = y[75:]

modelo = LinearSVC()
modelo.fit(treino_x,treino_y)

previsao = modelo.predict(teste_x)

acuracia = accuracy_score(teste_y, previsao) * 100
print(f"Taxa de acerto = {acuracia:.2f}%")
```
